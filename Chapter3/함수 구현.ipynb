소프트맥스 함수 구현

```python
import numpy as np
a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a)
print(exp_a)
```

```python
sum_exp_a = np.sum(exp_a)
print(sum_exp_a)
```

```python
y = exp_a / sum_exp_a
print(y)
```

```python
#함수로 만들기
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
#소프트맥스 오버플로
a = np.array([1010, 1000, 990])
np.exp(a) / np.sum(np.exp(a))
```

```python
#오버플로 막기위한 C 적용
c = np.max(a)
a - c
```

```python
np.exp(a - c) / np.sum(np.exp(a - c))
```

```python
#함수 구현
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a

    return y
```

소프트맥스 함수 특징

```python
#합은 1, 덕분에 확률로 해석가능
a = np.array([0.3, 2.9, 4.0])
y = softmax(a)
print(y)
np.sum(y)
```
