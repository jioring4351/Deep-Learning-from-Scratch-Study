신경망에서 행렬 곱

```python
#입력 2개(2), 가중치 각각 3개씩(2,3) = 3
X = np.array([1, 2])
X.shape
```

```python
W = np.array([[1,3,5], [2,4,6]])
print(W)
```

```python
W.shape
```

```python
Y = np.dot(X, W)
print(Y)
```
