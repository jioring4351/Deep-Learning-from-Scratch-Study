```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
#시그모이드 함수 그래프 실습
import matplotlib.pyplot as plt
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```
