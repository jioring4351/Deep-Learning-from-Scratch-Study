입력층(0층) 2개, 첫번째 은닉층(1층) 3개, 두번째 은닉층(2층) 2개, 출력층(3층) 2개


첫번째 층

```python
import numpy as np

X = np.array([1.0, 0.5]) #입력
W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]]) #첫번째 층 가중치
B1 = np.array([0.1, 0.2, 0.3]) #편향

print(W1.shape)
print(X.shape)
print(B1.shape)
```

```python
#첫번째 은닉층
A1 = np.dot(X, W1) + B1
```

```python
#첫번째 은닉층 활성화 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

Z1 = sigmoid(A1)

print(A1)
print(Z1)
```

두번째 층

```python
#첫번째 층의 출력 Z1이 2층의 입력이 된다
W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]]) #두번째 층 가중치
B2 = np.array([0.1, 0.2]) #편향

print(Z1.shape)
print(W2.shape)
print(B2.shape)
```

```python
#두번째 층 은닉층과 활성화 함수 적용
A2 = np.dot(Z1, W2) + B2
Z2 = sigmoid(A2)

print(A2)
print(Z2)
```

출력층(3층)

```python
#활성화 함수는 항등함수로
def identity_function(x):
    return x

W3 = np.array([[0.1, 0.3], [0.2, 0.4]]) #세번째 층 가중치
B3 = np.array([0.1, 0.2]) #편향

A3 = np.dot(Z2, W3) + B3
Y = identity_function(A3)

print(A3)
print(Y)
```

구현 정리

```python
def init_network():
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])

    return network

def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = identity_function(a3)

    return y

network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)
print(y)
```
